```python
import os
import json

def read_json_files_from_folder(folder_path):
    all_data = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r') as json_file:
                data = json.load(json_file)
                all_data += data  # Merge the lists directly
    return all_data

# Example usage:
folder_path = 'training_data/'
training_data = read_json_files_from_folder(folder_path)
for item in training_data:
    print(item)
```

```python
def convert_data(string, labels):
    # Split the string into words
    words = string.split()

    # Initialize the list to hold labels for each word
    word_labels = ['O'] * len(words)
    j = 0
    # Assign labels to words based on label_start and label_end
    for label_start, label_end, label_name in labels:
        for i, word in enumerate(words):
            # skip to next label
           
            """if j>i:
                continue"""
            # Calculate the cumulative length of words up to the current word
            cumulative_length = sum(len(w) + 1 for w in words[:i])  # Adding 1 for each space
            
            # Check if the label falls within the boundaries of the current word
            if label_start <= cumulative_length and label_end+1 >= cumulative_length + len(word):
                # Assign the label to all words within the range
                for j, _ in enumerate(range(i, len(words))):
                    cumulative_length = sum(len(w) + 1 for w in words[:j+i])
                    if cumulative_length + len(words[j+i]) <= label_end+1:
                        if word_labels[j+i] == "O":
                            word_labels[j+i] = 'B-' + label_name
                        else:
                            word_labels[j+i] = 'I-' + label_name
                    else:
                        break

    return words, word_labels

data = [['10 Crane, Samantha SO Moorpark 3 1:03.40', [[3, 17, 'ATHLETE_NAME'], [19, 21, 'YEAR'], [22, 29, 'SCHOOL'], [33, 39, 'SEED_TIME'], [0, 1, 'SEED_NUMBER']]],
[' 11 Reichle, Olivia SO Pomona-Pitzer 1 1:03.60', [[4, 18, 'ATHLETE_NAME'], [20, 21, 'YEAR'], [23, 28, 'SCHOOL'], [39, 45, 'SEED_TIME'], [1, 2, 'SEED_NUMBER']]],
[" 12 Dietrich, Rayanne FR Hope Int'l 3 1:03.90", [[4, 20, 'ATHLETE_NAME'], [22, 23, 'YEAR'], [25, 34, 'SCHOOL'], [38, 44, 'SEED_TIME'], [1, 2, 'SEED_NUMBER']]],
[' 13 Dawood, Razanne SO Orange Coast 2 1:04.30', [[4, 18, 'ATHLETE_NAME'], [20, 21, 'YEAR'], [23, 34, 'SCHOOL'], [38, 44, 'SEED_TIME'], [1, 2, 'SEED_NUMBER']]],
[' 14 Whitten, Rachel SO Saddleback 3 1:04.32', [[4, 18, 'ATHLETE_NAME'], [20, 21, 'YEAR'], [23, 32, 'SCHOOL'], [36, 42, 'SEED_TIME'], [1, 2, 'SEED_NUMBER']]]]

def convert_to_bert_data(data:list):
    words = []
    word_labels = []
    for line in data:
        w, wl = convert_data(line[0],line[1])
        words.append(w) # or extend?
        word_labels.append(wl)
    return words, word_labels

texts,tags = convert_to_bert_data(training_data)
# Print in a readable format
for item1, item2 in zip(texts, tags):
    print(item1, item2)
```

```python
from sklearn.model_selection import train_test_split
train_texts, val_texts, train_tags, val_tags = train_test_split(texts, tags, test_size=.2)
```

```python
unique_tags = set(tag for doc in tags for tag in doc)
tag2id = {tag: id for id, tag in enumerate(unique_tags)}
id2tag = {id: tag for tag, id in tag2id.items()}
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('dslim/distilbert-NER')
train_encodings = tokenizer(train_texts, is_split_into_words=True, return_offsets_mapping=True, padding=True, truncation=True)
val_encodings = tokenizer(val_texts, is_split_into_words=True, return_offsets_mapping=True, padding=True, truncation=True)
```

```python
import numpy as np

def encode_tags(tags, encodings):
    labels = [[tag2id[tag] for tag in doc] for doc in tags]
    encoded_labels = []
    for doc_labels, doc_offset in zip(labels, encodings.offset_mapping):
        # create an empty array of -100
        doc_enc_labels = np.ones(len(doc_offset),dtype=int) * -100
        arr_offset = np.array(doc_offset)

        # set labels whose first offset position is 0 and the second is not 0
        doc_enc_labels[(arr_offset[:,0] == 0) & (arr_offset[:,1] != 0)] = doc_labels
        encoded_labels.append(doc_enc_labels.tolist())

    return encoded_labels

train_labels = encode_tags(train_tags, train_encodings)
val_labels = encode_tags(val_tags, val_encodings)
```

```python
import torch

class WNUTDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_encodings.pop("offset_mapping") # we don't want to pass this to the model
val_encodings.pop("offset_mapping")
train_dataset = WNUTDataset(train_encodings, train_labels)
val_dataset = WNUTDataset(val_encodings, val_labels)
```

```python
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # total number of training epochs
    per_device_train_batch_size=16,  # batch size per device during training
    per_device_eval_batch_size=64,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)

model = DistilBertForSequenceClassification.from_pretrained("dslim/distilbert-NER",num_labels=len(unique_tags),ignore_mismatched_sizes=True)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

trainer = Trainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=val_dataset             # evaluation dataset
)

trainer.train()
eval_results = trainer.evaluate()
print(eval_results)
```
